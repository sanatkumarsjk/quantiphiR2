```python
import pandas as pd

data = pd.read_excel("train.xlsx")
data.head()
```

```python
# data.info()
# data.describe()
# labels = data[['Category','movie_sequel']]
# data = data.select_dtypes(exclude=['int64'])
data = data.drop(['name','display_name','board_rating_reason'], axis=1)
data = data.drop(['total', 'id', 'production_year'], axis=1)
data.info()
data = pd.get_dummies(data)
# data = pd.concat([data,labels],axis=1) 
print (data.columns)
data.describe()
```

```python
from sklearn.model_selection import train_test_split
data.head()
train_set, test_set = train_test_split(data, test_size=0.2, random_state=42)

trainF = train_set.drop("Category", axis=1)
trainL = train_set["Category"].copy()
testF = test_set.drop("Category", axis=1)
testL = test_set["Category"].copy()

trainF.shape
```

```python
# from sklearn.preprocessing import PolynomialFeatures
# print(trainF.shape)
# trainF = PolynomialFeatures(2).fit_transform(trainF)
# testF = PolynomialFeatures(2).fit_transform(testF)
# print(trainF.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score

clf = LinearRegression().fit(trainF, trainL)
print(clf.score(trainF, trainL))
print(clf.score(testF, testL))

y1_reg=clf.predict(testF)
y1_reg=list(y1_reg)

temp = []
for i in y1_reg:
    temp.append(int(round(i)))

correct = 0
for i,j in zip(testL.values, temp):
    if i == j or i+1 == j or i-1 == j:
        correct+=1
    
print(correct/len(testL))
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=0, solver='saga', multi_class='multinomial', max_iter=1000).fit(trainF, trainL)
print(clf.score(trainF, trainL))
print(clf.score(testF, testL))
y1_reg=clf.predict(testF)
y1_reg=list(y1_reg)

correct = 0
for i,j in zip(testL.values, y1_reg):
    if i == j or i+1 == j or i-1 == j:
        correct+=1
print(correct/len(testL))
```
